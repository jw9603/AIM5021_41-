### Counting Vocabulary(2th week)

```python
from nltk.book import *

# text1 출력
print(text1)

# length
print('text1의 길이',len(text1))

# 중복 제거후 token 오름차순 정렬
print('정렬',sorted(set(text1)))

# 고유한 token들 개수
print('number of tokens',len(set(text1)))

print('비율',len(set(text1))/len(text1))

# 'a' token의 count
print('count',text1.count('a'))

# count %
print(100*text1.count('a')/len(text1),'%')
```

### Counting Words by Genre

```python
from nltk.corpus import brown
import nltk
print(brown.categories())

genre_word = [(genre,word) for genre in ['news','adventure'] for word in brown.words(categories=genre)]
print(genre_word[:4])

print(len(genre_word))

cfd = nltk.ConditionalFreqDist(genre_word)

print(cfd['news'])

print(cfd['news'].most_common(5))

print(cfd['news']['the'])
```
